```python
# Linear graph

import matplotlib.pyplot as plt

plt.title("Basic Linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph a plot between distance and time

distance=list(range(2,11,2))
time=list(range(1,6))
#d=2*t
plt.title("Distance vs Time")
plt.xlabel("time")
plt.ylabel("distance")
plt.plot(time,distance,'go')
plt.grid()
plt.show()
```

```python
# v=a t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# plot a graph for number vs squares

n=list(range(10))
s=list(i**2 for i in n)
plt.plot(n,s)
plt.grid()
plt.show()
```

```python
# voltage vs time sinusoidal graph
import numpy as np
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.grid()
plt.xlabel('time')
plt.ylabel("voltage")
plt.title("Time Vs Voltage Graph")
plt.savefig("Images/voltage-time.png")
plt.show()

```

```python
# plot Income of Arkansas from 2005 to 2013

import pandas as pd
df=pd.read_csv("DataFiles/income.csv")
years=df.columns[2::]
income=df.values[3,2::]
plt.plot(years,income)
plt.xlabel('years')
plt.ylabel('income')
plt.title("Income of Arkansas")
plt.grid()
plt.show()
```

```python
#  


incomeallstates=[]
for i in df.columns[2::]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.title("Income of all states")
plt.xlabel("year")
plt.ylabel("Average Income")
plt.grid()
plt.show()
```
